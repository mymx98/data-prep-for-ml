# The Structure of matplotlib

The `matplotlib` is conceptually divided into three parts. The first part, matplotlib API, is the library that does the hard-work, managing and creating figures, text, lines, plots and so on. In the code above, we access this library by issuing the following command:

    >>> import matplotlib

The device dependent backend is the second part. It is the drawing engine that is responsible to render the visual representation to a file or a display device. Example of backends: 'PS' backend is used to create postscript file (suitable for hardcopy printout), 'SVG' creates scalar vector graphics (SVG file), 'Tkinter' on Windows provides interactive interface to the visualization. E.g. One can use the Agg backend to produce a PNG file, as displayed the example above:

     >>> matplotlib.use('Agg')

The `pyplot` interface is the last part of the matplotlib package. Module `pyplot` provides a set of functions that utilize the underlying matplotlib library. High level visualization functions like plot, boxplot, and bar, are available through pylab interface. To import these functions, issue the following command:

    >>> import matplotlib.pyplot as plt
    

```python
import matplotlib.pyplot as plt
```

## plot()
`Plot` is probably the most important function in matplotlib. 
`Plot` draws lines and/or markers using coordinates from the multiple points or x,y 
pairs supplied in the argument of the function. 
Both x and y are generally list or array of values. 

For example, the following command plots a simple quadratic function `y = x * x`.

    >>> x = [1,2,3,4]
    >>> y = [1,4,9,16]
    >>> plt.plot(x,y)

**Note:** A single list argument to the plot command, like `plot(y)`, would be considered as a list of y-values. `matplotlib` automatically generates the x-values for you.

**Note:** In shell script mode, everytime you issue `plot()` command, the output is added to the results of the earlier `plot()`. The `clf()` function can be called to empty the canvas. 



#### `figure, axes = plt.subplots()`

The code will create a figure with a single cell called `axes` 
where a plot will be drawn by `axes.plot()`

    this will directly draw a plot without creating a cell.

### Line plot
We will use the dataset we used in the lecture, `max_temp.csv`, for a multi-lines plot. 
Each line is a city's average maximum temperature for each month.

```python
import pandas as pd 

# create a DataFrame from the csv file
aus_temp = pd.read_csv('max_temp.csv')

aus_temp.head(10)
```

## line plots

### matplotlib
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
#This is a magic function that renders the figure in a notebook
%matplotlib inline
# create a figure with a single plot.
fig, ax = plt.subplots()

# plot a line avg max temperture for each city
for i, row in aus_temp.iterrows():
    ax.plot(row[1:])
    
ax.legend(aus_temp['city/month'])
```

## Histograms

### matplotlib
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
# create a figure with a single plot.
fig, ax = plt.subplots()

ax.hist(car['horsepower'])
```

```python
bin_sizes = [3, 10, 30]

fig, axs = plt.subplots(3)
fig.suptitle('Histograms with different bin sizes')

for i, size in enumerate(bin_sizes):
    axs[i].hist(car['horsepower'], bins=size)
```

## Scatter plots

### matplotlib
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
# we have 3 subsets of cars, from 3 different origin. We are interested in exploring the relationships between horsepower and acceleration.
car1 = car.loc[car['origin'] == 1, ['horsepower', 'acceleration']]
car2 = car.loc[car['origin'] == 2, ['horsepower', 'acceleration']]
car3 = car.loc[car['origin'] == 3, ['horsepower', 'acceleration']]

car1.shape, car2.shape, car3.shape
```

```python
# create a figure with a single plot.
fig, ax = plt.subplots()

ax.scatter(car1['horsepower'],car1['acceleration'],color='purple')
ax.scatter(car2['horsepower'],car2['acceleration'],color='teal')
ax.scatter(car3['horsepower'],car3['acceleration'],color='yellow')
```

### pandas
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.scatter.html

```python
car.plot.scatter(x='horsepower', y='acceleration', c='origin', colormap='viridis')
```

## Parallel co-ordinate plots

https://pandas.pydata.org/docs/reference/api/pandas.plotting.parallel_coordinates.html

`>>> from pandas.plotting import parallel_coordinates`

```python
from pandas.plotting import parallel_coordinates 
```

```python
# prepare the data
columns = car.columns.tolist()
# select the numeric features
features = car[columns[:6]].copy()
# process class labels
features['class'] = ['origin ' + str(x) for x in car['origin']]

```

```python
parallel_coordinates(features, class_column = 'class', color = ['blue', 'green', 'orange'])
```

```python
# normalise the data
columns = car.columns.tolist()
features = car[columns[:6]].copy()

features_norm = (features - features.min())/(features.max() - features.min())
# features_sd = (features - features.mean())/features.std()
features_norm['class'] = ['origin ' + str(x) for x in car['origin']]
features_norm
```

```python
parallel_coordinates(features_norm, class_column = 'class', color = ['blue', 'green', 'orange'])
```

## Heatmaps

### matplotlib
One way to draw a heatmap of a matrix is matshow()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.matshow.html#matplotlib.pyplot.matshow
    

```python
# create a figure with a single plot.
fig, ax = plt.subplots()
ax.matshow(features, aspect='auto')
```

### seaborn
The seaborn package provides very nice function for drawing a heatmap.

https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

features['class'] = ['origin ' + str(x) for x in car['origin']]
features = features.set_index('class').sort_index()
fig, ax = plt.subplots()
sns.heatmap(features, cmap='viridis', xticklabels=True, ax=ax)
ax.tick_params(axis='x', labelrotation=30, direction='out')
```

```python

features_norm = (features - features.min())/(features.max() - features.min())
#features_sd = (features - features.mean())/features.std()

features_norm['class'] = ['origin ' + str(x) for x in car['origin']]
features_norm = features_norm.set_index('class').sort_index()

fig, ax = plt.subplots()
sns.heatmap(features_norm, cmap='viridis', xticklabels=True, ax=ax)
```
